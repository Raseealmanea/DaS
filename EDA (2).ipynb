```python

correlation_matrix = df[['Views', 'Likes', 'Comments', "Duration"]].corr()


plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="Blues", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()


```

Views have a strong positive correlation with Likes (0.78) and a moderate correlation with Comments (0.49). This suggests that as the number of views increases, the number of likes also tends to increase significantly, indicating that videos with higher view counts are likely to engage viewers and receive more likes. The moderate correlation with comments implies that while more views generally lead to more comments, the relationship is not as strong as with likes, indicating that viewers may be more likely to express approval through likes rather than comments.

Likes exhibit a very strong positive correlation with Comments (0.79). This indicates that when viewers like a video, they are also more inclined to leave a comment. The strong relationship between these two metrics suggests that videos that resonate well with audiences will not only receive likes but also prompt further interaction in the form of comments.

Conversely, the relationship between video Duration and the other metrics (Views, Likes, Comments) is negligible. this  may indicate that longer videos tend to attract fewer views, likes, and comments, implying that audiences may prefer shorter videos, which can negatively impact overall engagement.

How does the use of a particular language affect interaction?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# extract column of language
language_cols = [col for col in df.columns if "Language_" in col]


language_views = df[language_cols].multiply(df["Views"], axis=0).sum()
language_comments = df[language_cols].multiply(df["Comments"], axis=0).sum()
language_likes = df[language_cols].multiply(df["Likes"], axis=0).sum()


language_stats = pd.DataFrame({
    "Views": language_views,
    "Comments": language_comments,
    "Likes": language_likes
})


plt.figure(figsize=(12, 6))
language_stats.plot(kind='bar', figsize=(12, 6), color=['skyblue', 'lightpink', 'lightgreen'])

plt.title("Comparison of Views, Comments, and Likes per Language")
plt.xlabel("Language")
plt.ylabel("Total Count of views, comment and like")
plt.xticks(rotation=90)
plt.legend(["Views", "Comments", "Likes"])
plt.show()
```

From the graph, we can see that the use of a specific language can significantly increase interaction by raising the number of views, likes, and comments. It is evident that the English language accounts for the highest levels of interaction, featuring a substantial number of views, likes, and comments. Following English, the Spanish language also shows considerable engagement, though it is lower than that of English.

This indicates that the choice of language plays a crucial role in the reach and interaction of content. Additionally, the graph reveals that languages like Russian and Hindi have much lower interaction levels. This may reflect various factors such as audience size and cultural relevance, which can affect how content is received and engaged with by different demographic groups.

In summary, the graph illustrates that content in widely spoken languages, particularly English and Spanish, tends to attract more interaction, while content in languages with smaller audiences, such as Russian and Hindi, may face challenges in achieving similar engagement levels.

 What are the most popular topics in Saudi Arabia?

```python

df_saudi = df[df["Region"] == "SA"]


topic_cols = [col for col in df_saudi.columns if "Topic_" in col]


topic_counts = df_saudi[topic_cols].sum()


plt.figure(figsize=(12, 5))
topic_counts.sort_values(ascending=True).plot(kind='barh', color='lightblue')
plt.title("Most Popular Topics in Saudi Arabia")
plt.xlabel("Number of Videos")
plt.ylabel("Topic")
plt.show()
```

From the graph, the most popular topics in Saudi Arabia are:

1. Entertainment
2. Gaming
3. Sports
4. People & Blogs
5. Music

Overall, the graph highlights that entertainment-focused topics, particularly gaming and sports, dominate the landscape in Saudi Arabia.

Is there a relationship between posting time (morning/evening) and number of views?

```python

df['Publication Hour'] = pd.to_datetime(df['Publication Time']).dt.hour

df['Time of Day'] = df['Publication Hour'].apply(lambda x: 'Morning' if 6 <= x < 18 else 'Evening')

views_pie = df.groupby("Time of Day")["Views"].sum()

plt.figure(figsize=(7, 7))
views_pie.plot(kind='pie', autopct='%1.1f%%', colors=['blue', 'lightblue'])
plt.title("Percentage of Views by Posting Time")
plt.ylabel("")
plt.show()
```

Based on the pie chart, there appears to be a relationship between posting time and the number of views. Specifically, 56.6% of the posts were made in the morning, while 43.4% were made in the evening. This suggests that posts made in the morning receive a higher proportion of views compared to those posted in the evening.

 How does the duration of the video affect the number of views or interaction?

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Duration"], y=df["Views"], alpha=0.5, color='purple')
plt.title("Video Duration vs. Views")
plt.xlabel("Duration (Seconds)")
plt.ylabel("Views")
plt.show()
```

Based on the scatter plot comparing video duration (in seconds) to the number of views, we can deduce several important points. Firstly, there is an inverse relationship between video length and views, as shorter videos (under 10,000 seconds) tend to receive higher view counts, while longer videos (over 20,000 seconds) generally attract fewer views. Most data points cluster around low view counts, indicating that videos exceeding a certain duration do not gain adequate engagement. Additionally, as video duration increases, there is a clear decline in the number of views, suggesting that longer content may negatively impact viewer interest.
